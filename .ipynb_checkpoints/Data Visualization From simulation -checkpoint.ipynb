```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import Markdown, display
def printmd(string):
    display(Markdown(string))
```

loading the "big" datafile

```python
df = pd.read_csv('big.csv')
df.head()
```

```python
df['sim_mode'] = df.CSCS.str.split("-").str.get(0)
df['sim_wavelength'] = df.CSCS.str.split("-").str.get(1)
df['latice_basis'] = df.CSCS.str.split("-").str.get(2).str.split("/").str.get(0)
df['material']= df.CSCS.str.split("-").str.get(2).str.split("/").str.get(1).str.split('=').str.get(0)
df['material_thickness'] = df.CSCS.str.split("-").str.get(2).str.split("/").str.get(1).str.split('=').str.get(1).str.split(':').str.get(0)
df['patter_shape'] = test = df.CSCS.str.split("-").str.get(2).str.split("/").str.get(1).str.split('=').str.get(1).str.split(':').str.get(1).str.get(0)
Temp = df.CSCS.str.split("-").str.get(2).str.split("/").str.get(1).str.split('=').str.get(1).str.split(':').str.get(1).str[1::]
Temp = Temp.str.replace("(","").str.replace(")","")
df['pattern_positionX'] = Temp.str.split(',').str[0]
df['pattern_positionY'] = Temp.str.split(',').str[1]
df['radius'] = Temp.str.split(',').str[2]

cleanData = df.loc[:, df.columns != 'CSCS']
cleanData.head()
```

```python
for col in cleanData.drop(['transmission','phase'], axis=1):
    
    printmd("for **%s**: number of unique values are: %d" % (col,cleanData[col].nunique()))
    #print(cleanData[col].unique())

wavelengths = set(cleanData['sim_wavelength'])
sorted_wavelength = sorted(wavelengths)

wavel1 = cleanData[cleanData['sim_wavelength']==sorted_wavelength[0]]
wavel2 = cleanData[cleanData['sim_wavelength']==sorted_wavelength[1]]
wavel3 = cleanData[cleanData['sim_wavelength']==sorted_wavelength[2]]

for col in wavel1.drop(['transmission','phase'], axis=1):
    
    printmd("for Wavelength **%s** **%s**: number of unique values are: %d" % (sorted_wavelength[0],col,wavel1[col].nunique()))
    #print(cleanData[col].unique())

for col in wavel2.drop(['transmission','phase'], axis=1):
    
    printmd("for Wavelength **%s** **%s**: number of unique values are: %d" % (sorted_wavelength[1],col,wavel2[col].nunique()))
    #print(cleanData[col].unique())

for col in wavel3.drop(['transmission','phase'], axis=1):
    
    printmd("for Wavelength **%s** **%s**: number of unique values are: %d" % (sorted_wavelength[2],col,wavel3[col].nunique()))
    #print(cleanData[col].unique())
```
